# 作業:

 當 "projection" 參數為“Mill"時，所得平展位面

```python
# 載入套件
import os
os.environ["PROJ_LIB"] = "D:\\Anaconda3\\Library\\share"
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 創建基本地圖。
# 當 "projection" 參數為“ortho"時，所得圖位地球儀截面
'''
#由此開始寫code 

#設定圖形大小
plt.figure(...)

#設定投影型態
m=Basemap(....)

#設定圖形顏色, 選用NASA 寶石藍輸出 
m.bluemarble()

#秀圖
plt.show()
'''

plt.figure(figsize=(5,5)) #figsize的單位為英寸
m=Basemap(projection='ortho',resolution = None, lat_0=23,lon_0=120)
### Default image size = 5400x2700, thus scale to save memory ###
m.bluemarble(scale=0.25) 
plt.show()
```

```python
'''
由此開始寫code 

#設置Lambert保形底圖。
# width=12000000,height=9000000
#經緯度lat_1=45.,lat_2=55,lat_0=50,lon_0=-107
#set resolution = None跳過邊界數據集的處理
# projection='lcc'

m = Basemap(....)

#選用NASA 寶石藍輸出 
m.bluemarble()

#秀圖
plt.show()
'''
m1 = Basemap(projection ='lcc', width=12000000, height=10000000, lat_1=45.,lat_2=55,lat_0=50,lon_0=-107, resolution = None)
m1.bluemarble(scale=0.25)
plt.show()
```

```python
'''
由此開始寫code 
#  "projection" 參數為“Mill"
# 設定經緯座標
# 左下角的緯度 - llcrnrlat = -90,
# 左下角的經度 - llcrnrlon = -180,
# 右上角的緯度 - urcrnrlat = 90,
# 右上角的經度 -urcrnrlon = 180

# 成圖設定resolution

m = Basemap() # 成圖設定resolution 

#創建邊線
m.drawcoastlines()
#畫出國家，並使用線寬為2 的線條生成分界線。
m.drawcountries()
#畫出州界
m.drawstates(color='b')
#畫出城市
m.drawcounties(color='darkred')

plt.title('Basemap Tutorial')
plt.show()

'''
m2= Basemap(projection = 'mill', llcrnrlat = -90,llcrnrlon = -180,urcrnrlat = 90,urcrnrlon = 180, resolution ='c') # 成圖設定resolution 

#創建邊線
m2.drawcoastlines()
#畫出國家，並使用線寬為2 的線條生成分界線。
m2.drawcountries(linewidth=2)
#畫出州界
m2.drawstates(color= 'b')
#畫出城市 --未安裝counties圖
#m2.drawcounties(color= 'r')

plt.title('Basemap Tutorial')
plt.show()
```
